## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

# Path is relative to work/notebooks/, pointing at the real starter dataset
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

n_pages = len(df)
pct_declining = (df["trend_direction"] == "down").mean() * 100
avg_position = df["avg_position"].mean()
avg_ctr = df["ctr"].mean()
median_ctr = df["ctr"].median()

print(f"Pages in the anonymized starter dataset: {n_pages:,}")
print(f"Share with a declining traffic trend: {pct_declining:.1f}%")
print(f"Average search position: {avg_position:.1f}")
print(f"Average CTR: {avg_ctr:.2f}%  |  Median CTR: {median_ctr:.2f}%")
```

**What these numbers say**

Across the 30,000-page anonymized starter set, just over half the pages (about 54%) have a declining traffic trend, and the average search position sits around 16 — page two territory, not the top of page one. Average CTR is only about 0.5%, and the median page's CTR is closer to 0.1% — most pages are visible in search but barely get clicked. That gap between being indexed/ranked and actually earning clicks is the observable pattern this lane is built on: it's not something I'm assuming, it shows up directly in the data before any model touches it. These are all observed metrics, not the pipeline's own scores — I'm deliberately not using derived fields like a refresh score or a model's confidence label to justify the lane, since that would be circular.
